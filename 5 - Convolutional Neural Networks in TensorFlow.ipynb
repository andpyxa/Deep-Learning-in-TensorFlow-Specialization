## 5.1 Neural Network and Convolutional Neural Networks


### <font color='Orange'> Convolutional Neural Network </font>

<font size="3">**A convolutional neural network, is a network architecture for deep learning which learns directly from data, eliminating the need for manual feature extraction.**</font> 

<font size="3">**Convolutional neural networks have had great success in certain kinds of problems, such as image recognition. They are particularly useful for finding patterns in images to recognize objects, faces, and scenes. They can also be quite effective for classifying non-image data such as audio, time series, and signal data.**</font> 

<font size="3">**By definition, a <span style="color:#4285F4">neural network</span> in which at least one layer is a <span style="color:#4285F4">convolutional layer</span> is defined as <span style="color:#4285F4">convolutional neural network</span>.**</font>
    
<font size="3">**A typical <span style="color:#4285F4">convolutional neural network</span> consists of some combination of the following layers:**</font>

><font size="3">**<span style="color:#4285F4">Convolutional layers</span>**</font><br>

><font size="3">**<span style="color:#4285F4">Pooling layers</span>**</font><br>

><font size="3">**<span style="color:#4285F4">Dense layers</span>**</font><br>

### <font color='Orange'> Convolutional Layer </font>

<font size="3">**A layer of a <span style="color:#4285F4">deep neural network</span> in which a <span style="color:#4285F4">convolutional filter</span> passes along an input matrix.**</font>
    
<font size="3">**For example, there is a <span style="color:red">3x3</span> <span style="color:#4285F4">convolutional filter (kernel)</span>**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNNKernel.png?alt=media&token=01640c6d-166f-4726-b963-dfc41e0a79ea" width="150" style="float"/> 

<font size="3">**Let's say we have a <span style="color:red">5x5</span> input matrix. Then, the <span style="color:#4285F4">convolutional filter</span> will consist of <span style="color:red">9</span> convolutional operations.**</font> 

<font size="3">**Each convolutional operation works on a different <span style="color:red">3x3</span> slice of the input matrix. The resulting <span style="color:red">3x3</span> matrix (on the right) consists of the results of the <span style="color:red">9</span> convolutional operations:**</font>

### <font color='Orange'> Why Convolutional Neural Network? </font>

<font size="3">**Using convolutional neural networks for deep learning is popular due to three important factors:**</font>

><font size="3">**It <span style="color:#4285F4">eliminates the need for manual feature extraction</span> â€” the features are learned directly by the convolutional neural networks.**</font>

><font size="3">**It produces <span style="color:#4285F4">highly accurate</span> recognition results.**</font>

><font size="3">**It cans be retrained for new recognition tasks, enabling you to <span style="color:#4285F4">build on pre-existing networks</span>.**</font>

<font size="3">**An image is nothing but a matrix of pixel values. So it is always possible to flatten the image (e.g. 3x3 image matrix into a 9x1 vector) and feed it to a neural network for classification purposes.**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN1.png?alt=media&token=2b46bb8f-0ceb-46bc-b4af-1896d9989627" width="200" style="float:center"/>

<font size="3">**In cases of extremely basic binary images, the method might be fine. However, when it comes to complex images having pixel dependencies throughout, it would have little to no accuracy.**</font>

<font size="3">**Therefore, another method is needed. A <span style="color:#4285F4">convolutional neural network</span> is able to successfully capture the spatial and temporal dependencies in an image through the application of <span style="color:#4285F4">convolutional filters</span>.**</font>
    
<font size="3">**This architecture provides an optimal architecture for uncovering and learning key features in image and time-series data. It <span style="color:#4285F4">reduces the amount of neurons (weights)</span> significantly and <span style="color:#4285F4">avoid overfitting</span> the model. In addition, it keeps the reusability of neurons (weights). As a result, the network can be trained to understand the sophistication of an image better.**</font>

## 5.2 Basic Architecture of Convolutional Neural Networks

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN2.png?alt=media&token=9b2ffe9a-bbe2-40e9-986f-2f47ef012b91" width="950" align="center"/>

### <font color='Orange'> How Convolutional Neural Network works? </font>

### <font color='#34A853'> Feature Learning </font>

<font size="3">**A convolutional neural network can have tens or hundreds of layers that each learn to detect different features of an image.**</font>

<font size="3">**<span style="color:#4285F4">Filters</span> are applied to each training image at different resolutions, and the output of each convolved image is used as the input to the next layer.**</font> 
    
<font size="3">**The filters can <span style="color:#4285F4">start as very simple features</span>, such as brightness and edges, and <span style="color:#4285F4">increase in complexity to features</span> that uniquely define the object.**</font>

### <font color='#34A853'> Layers </font>

<font size="3">**Like other neural networks, a convolutional neural network is composed of**</font> 

><font size="3">**<span style="color:#4285F4">an input layer</span>**</font>

><font size="3">**<span style="color:#4285F4">an output layer</span>**</font>

><font size="3">**<span style="color:#4285F4">any hidden layers in between</span>**</font>

<font size="3">**These layers perform operations that alter the data with the intent of learning features specific to the data. Three of the most common layers are:**</font> 

><font size="3">**<span style="color:#4285F4">convolution</span>** - this puts the input images through a set of <span style="color:#4285F4">convolutional filters</span>, each of which activates certain features from the images.</font>

><font size="3">**<span style="color:#4285F4">activation</span>** - a good combination allows for faster and more effective training because only the activated features are carried forward into the next layer</font>

><font size="3">**<span style="color:#4285F4">pooling</span>** - this simplifies the output by performing nonlinear downsampling, reducing the number of parameters that the network needs to learn</font>

<font size="3">**Convolution puts the input images through a set of <span style="color:#4285F4">convolutional filters</span>, each of which activates certain features from the images.**</font>

<font size="3">**These operations are repeated over tens or hundreds of layers, with each layer learning to identify different features.**</font>

### <font color='#34A853'> Shared Weights and Biases </font>

<font size="3">**Like a traditional neural network, a convolutional neural network has neurons with weights and biases. The model learns these values during the training process, and it continuously updates them with each new training example.**</font>

<font size="3">**However, in the case of convolutional neural networks, the <span style="color:#4285F4">weights and bias values are the same for all hidden neurons in a given layer</span>. This arrangement ensures that**</font>

><font size="3">**all hidden neurons are detecting the <span style="color:#4285F4">same feature</span>, such as an edge or a blob, in different regions of the image.**</font>
    
><font size="3">**the network is tolerant to translation of objects in an image. E.g., a network trained to recognize dogs will be able to do so wherever the dog is in the image.**</font>

### <font color='#34A853'> Classification </font>

<font size="3">**After <span style="color:#4285F4">learning features</span> in many layers, the architecture of a convolutional neural network shifts to <span style="color:#4285F4">classification</span>.**</font>

<font size="3">**The <span style="color:#4285F4">next-to-last</span> layer is a <span style="color:#4285F4">fully connected layer</span> that outputs a vector of classes that the network will be able to predict. This vector contains the <span style="color:#4285F4">probabilities</span> for each class of any image being classified.**</font>

<font size="3">**The final layer of the convolutional neural network architecture uses a <span style="color:#4285F4">classification layer</span> such as <span style="color:#4285F4">softmax</span> to provide the classification output.**</font>

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

### <font color='#34A853'>Preprocess the data - Normalization</font>
<hr style="border:2px solid ##34A853"> </hr>

<font size="3">**The images are stored in 28x28 numpy arrays, with pixel values ranging from 0 to 255. Therefore, value 255 is used to normalize pixel values to be between 0 and 1.**</font>   

```python
train_images, test_images = train_images / 255.0, test_images / 255.0
```

<font size="3">**The labels are an array of integers, ranging from 0 to 9. Their corresponding class names can be stored in a list.**</font>   

```python
class_names = ['airplane', 
               'automobile', 
               'bird', 
               'cat', 
               'deer',
               'dog', 
               'frog', 
               'horse', 
               'ship', 
               'truck']
```

### <font color='#34A853'>Explore the data</font>

```python
print("The shape of training images is:", train_images.shape)
print("The shape of training labels is:", train_labels.shape)
```

### <font color='#34A853'>Build the model</font>
<hr style="border:2px solid ##34A853"> </hr>

### <font color='Orange'>*1) Setup and configure a Sequential model*</font>



<font size="3">**As input, a convolutional neural network takes tensors of shape (<span style="color:#4285F4">image_height</span>, <span style="color:#4285F4">image_width</span>, <span style="color:#4285F4">color_channels</span>), ignoring the batch size. color_channels refers to (R,G,B).**</font> 
    
<font size="3">**In this example, convolutional neural network is configured to process inputs of shape <span style="color:#4285F4">(32, 32, 3)</span>, which is the format of CIFAR images. <span style="color:#4285F4">input_shape</span> argument can be set in first layer.**</font>

<font size="3"><u>**Create the convolutional base**</u></font> 

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN4.png?alt=media&token=9a22a887-e79d-477d-98e8-05ff40307309" width="950" align="center"/>

<font size="3">**To create the feature learning layers as the convolutional base, let's use a common pattern:**</font>
><font size="3">**a stack of <span style="color:#4285F4">Conv2D</span> and <span style="color:#4285F4">MaxPooling2D</span> layers**</font>

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
```

```python
model.summary()
```

<font size="3">**The output of every <span style="color:#4285F4">Conv2D</span> and <span style="color:#4285F4">MaxPooling2D</span> layer is a 3D tensor of shape (height, width, channels).**</font>

><font size="3">**The width and height dimensions tend to shrink as it goes deeper in the network.**</font>

><font size="3">**The number of output channels for each <span style="color:#4285F4">Conv2D</span> layer is controlled by the number of filters, i.e. first argument (e.g., 32 or 64).**</font>

<font size="5"><span style="background-color:#EA4335; color:white">&nbsp;!&nbsp;</span></font> <font size="3">**Typically, as the width and height shrink, more computational resource is affordable. Hence, more output channels can be added in each <span style="color:#4285F4">Conv2D</span> layer.**</font>

<font size="3"><u>**Add classification on top of the convolutional base**</u></font> 

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN5.png?alt=media&token=17de648f-c844-4406-ba24-da83ce490e9d" width="950" align="center"/>

<font size="3">**To complete the model, one or more <span style="color:#4285F4">dense layers</span> can be added on top of the last output tensor from the convolutional base (of shape (4, 4, 64)) to perform classification.**</font>
    
<font size="3">**However, <span style="color:#4285F4">dense layers</span> take <span style="color:#4285F4">vectors</span> as input (which are 1D), while the current output is a <span style="color:#4285F4">3D tensor</span>.**</font>

<font size="3">**Therefore,**</font>
><font size="3">**firstly, the 3D output needs to be flattened (or unrolled) to 1D with the use of <span style="color:#4285F4">flatten layers</span>**</font><br>

><font size="3">**then one or more <span style="color:#4285F4">dense layers</span> can be further added,**</font><br>
    
><font size="3">**then a final <span style="color:#4285F4">dense layer</span> is added with <span style="color:#4285F4">10 outputs</span> to fit the CIFAR's <span style="color:#4285F4">10 output classes</span>.**</font><br>


```python
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation ='softmax'))
```

<font size="3">**Here's the complete architecture:**</font><br>

```python
model.summary()
```

<font size="3">**At the end of the convolutional base, outputs (4, 4, 64) are flattened into vectors of shape (1024), then are fed through two dense layers.**</font><br>

### <font color='Orange'>*2) Compile the model*</font>

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['accuracy'])
```

<font size="5"><span style="background-color:#EA4335; color:white">&nbsp;!&nbsp;</span></font> <font size="3">**Remark: If <span style="background-color: #ECECEC; color:#0047bb">from_logits</span> is <span style="color:#4285F4">True</span>, no activation function is needed in the output layer, e.g.**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/3NNExercise2.png?alt=media&token=d053ce8a-8ccb-4087-a198-a19ced62e298" width="2000"/>

### <font color='#34A853'>Train the model</font>
<hr style="border:2px solid ##34A853"> </hr>

### <font color='Orange'>*1) Feed the model*</font>

```python
history = model.fit(train_images, train_labels, epochs=10, 
                    validation_data=(test_images, test_labels))
```

<font size="3">**To start training, <span style="background-color: #ECECEC; color:#0047bb">model.fit()</span> method can be used to "fit" the model with training data. In this example, the training data is in the <span style="color:#4285F4">train_images</span> and <span style="color:#4285F4">train_labels</span> arrays.**</font>

<font size="3">***Training accuracy***</font>
> <font size="3">**As the model trains, the loss and accuracy metrics are improving. After 10 epochs, this model reaches an accuracy of about 0.79 (79%) on training data.**</font><br>

### <font color='Orange'>*2) Evaluate the model*</font>

```python
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.3, 1])
plt.legend(loc='lower right')
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)

print('Test accuracy:', round(test_acc*100), '%')
```

<font size="3">***Test accuracy***</font>
> <font size="3">**It turns out this convolutional neural network has achieved a test accuracy of around 70%. Not bad with such simple model!**</font>

### <font color='Orange'>*3) Verify and visualize predictions*</font>

<font size="3">**With trained network, it can be used to make predictions.**</font>
<font size="3">**Let's predict 16 images. Correct prediction labels are blue and incorrect prediction labels are red. The number indicates the precentage (i.e. confidence) for the predicted label.**</font>

```python
predictions = model.predict(test_images)

plt.figure(figsize=(17,8))

for i in range(16):
    plt.subplot(4,8,2*i+1)
    plt.xticks([])
    plt.yticks([])
    if np.argmax(predictions[i]) == test_labels[i][0]:
        color = 'blue'
    else:
        color = 'red'
    plt.imshow(test_images[i], cmap=plt.cm.binary)
    plt.xlabel("{} {:2.0f}% ({})".format(class_names[np.argmax(predictions[i])],
                                100*np.max(predictions[i]),
                                class_names[test_labels[i][0]]),
                                color=color)
    plt.subplot(4,8,2*i+2)
    plt.xticks(range(10))
    plt.yticks([])
    thisplot = plt.bar(range(10), predictions[i], color="grey")
    plt.ylim([0, 1])  

    thisplot[np.argmax(predictions[i])].set_color('red')
    thisplot[test_labels[i][0]].set_color('blue')
    
plt.show()
```

<font size="5"><span style="background-color:#EA4335; color:white">&nbsp;!&nbsp;</span></font> <font size="3">**The model can be wrong even when it has high confidence about its prediction.**</font>

## 5.3 Convolutional Layer

<font size="3">**There are several drawbacks of neural networks, especially when it comes to image processing.**</font>

><font size="3">**<span style="color:#4285F4">Numerous amount of weights</span>**</font><br>
><font size="3"></font><br>
><font size="3">**Neural networks use one neural for each input. The amount of weights rapidly becomes unmanageable for large images. For a 200 x 200 pixel image with 3 color channels there are around 120,000 weights that must be trained. As a result, difficulties arise whilst training and overfitting can occur.**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN1.png?alt=media&token=2b46bb8f-0ceb-46bc-b4af-1896d9989627" width="200" style="float:center"/>


### <font color='#176BEF'> Examples </font>
<hr style="border:2px solid #E1F6FF"> </hr>

<font size="3">**If we use neural network to process an 32x32x3 image, it ends up we will have <span style="color:red">197,322</span> weights.**</font>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model_NN = Sequential()
model_NN.add(Flatten(input_shape=(32,32,3)))
model_NN.add(Dense(64, activation ='relu'))
model_NN.add(Dense(10, activation ='softmax'))
model_NN.summary()
```

<font size="3">**With the use of <span style="color:#4285F4">Convolutional Layers</span>, we manage to have <span style="color:red">122,570</span> weights.**</font>

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D((2, 2)))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation ='softmax'))
model.summary()
```

### <font color='Orange'> Convolution </font>

<font size="3">**Instead, a convolutional neural network will use a mathematical technique, namely <span style="color:#4285F4">convolution</span>, to extract only the most relevant pixels. This technique allows the network to learn increasingly <span style="color:#4285F4">complex features</span> at each layer.**</font>

><font size="3">**The <span style="color:#4285F4">convolution</span> divides the matrix into small pieces to learn to most essential elements within each piece.**</font>

><font size="3">**In other words, <span style="color:#4285F4">convolution</span> allows the model to learn specific patterns within the picture and will be able to recognize it everywhere in the picture.**</font>

<font size="3">**<span style="color:#4285F4">Convolution</span> is an <span style="color:#4285F4">element-wise multiplication</span>. It will scan a part of the image, usually with a dimension of <span style="color:#4285F4">3Ã—3 window</span> and multiplies it to a <span style="color:#4285F4">filter (kernel)</span>.**</font>

<font size="3">**The output of the element-wise multiplication is called a <span style="color:#4285F4">feature map</span>. This step is repeated until all the image is scanned.**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNNgif1.gif?alt=media&token=f9245b7e-104f-43a0-bc73-c20fa76383f5" width="530" style="float:center"/>

### <font color='#34A853'>Kernel</font>

><font size="3">**Convolution applies a <span style="color:#4285F4">kernel</span> to extract certain <span style="color:#4285F4">features</span> from an input image.**</font>

><font size="3">**In general, <span style="color:#4285F4">small</span> kernel sizes are preferred over fully connected network due to computational costs. In addition, <span style="color:#4285F4">odd-sized</span> kernels are preferred because it symmetrically divide the previous layer pixels around the output pixel. Therefore, <span style="color:#4285F4">3x3</span> is a very popular and optimal choice.**</font>

<font size="3">**This graph shows how kernel works for the first element:**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN11.png?alt=media&token=a9911e97-8ae4-4a2a-af34-088643401b54" width="750" style="float:center"/>

><font size="3">**Depending on the kernel's <span style="color:#4285F4">element values</span>, a kernel can serve different purposes, such as <span style="color:#4285F4">edge detection</span>, <span style="color:#4285F4">sharpen</span> and <span style="color:#4285F4">box blur</span>.**</font>

```python
from IPython.display import HTML
from IPython.display import display

tag = HTML('''<script>
code_show=true; 
function code_toggle() {
    if (code_show){
        $('div.cell.code_cell.rendered.selected div.input').hide();
    } else {
        $('div.cell.code_cell.rendered.selected div.input').show();
    }
    code_show = !code_show
} 
$( document ).ready(code_toggle);
</script>
<a href="javascript:code_toggle()">Show or Hide the code</a>''')
display(tag)

#################################

from skimage import io
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt

def convolve2D(image, kernel, padding=0, strides=1):
    # Cross Correlation
    kernel = np.flipud(np.fliplr(kernel))

    # Gather Shapes of Kernel + Image + Padding
    xKernShape = kernel.shape[0]
    yKernShape = kernel.shape[1]
    xImgShape = image.shape[0]
    yImgShape = image.shape[1]

    # Shape of Output Convolution
    xOutput = int(((xImgShape - xKernShape + 2 * padding) / strides) + 1)
    yOutput = int(((yImgShape - yKernShape + 2 * padding) / strides) + 1)
    output = np.zeros((xOutput, yOutput))

    # Apply Equal Padding to All Sides
    if padding != 0:
        imagePadded = np.zeros((image.shape[0] + padding*2, image.shape[1] + padding*2))
        imagePadded[int(padding):int(-1 * padding), int(padding):int(-1 * padding)] = image
    else:
        imagePadded = image

    # Iterate through image
    for y in range(image.shape[1]):
        # Exit Convolution
        if y > image.shape[1] - yKernShape:
            break
        # Only Convolve if y has gone down by the specified Strides
        if y % strides == 0:
            for x in range(image.shape[0]):
                # Go to next row once kernel is out of bounds
                if x > image.shape[0] - xKernShape:
                    break
                try:
                    # Only Convolve if x has moved by the specified Strides
                    if x % strides == 0:
                        output[x, y] = (kernel * imagePadded[x: x + xKernShape, y: y + yKernShape]).sum()
                except:
                    break
    return output

image = io.imread('https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN12.png?alt=media&token=544238ec-eec0-4638-9d0f-453ea97aaf5d')
image = image[:,:,0:3]

grayscale = rgb2gray(image)

image_edge_detection = convolve2D(grayscale, np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]), padding=1, strides=1)
image_sharpen = convolve2D(grayscale, np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]), padding=1, strides=1)
image_blur = convolve2D(grayscale, np.array([[1/9, 1/9, 1/9], [1/9, 1/9, 1/9], [1/9, 1/9, 1/9]]), padding=1, strides=1)

plt.figure(figsize=(17,17))
plt.subplot(1,4,1)
plt.xticks([])
plt.yticks([])
plt.imshow(grayscale, cmap='gray')
plt.title('Original Image', fontsize=18)

plt.subplot(1,4,2)
plt.xticks([])
plt.yticks([])
plt.imshow(image_edge_detection, cmap='gray')
plt.title('Edge Detection', fontsize=18)

plt.subplot(1,4,3)
plt.xticks([])
plt.yticks([])
plt.imshow(image_sharpen, cmap='gray')
plt.title('Sharpen', fontsize=18)

plt.subplot(1,4,4)
plt.xticks([])
plt.yticks([])
plt.imshow(image_blur, cmap='gray')
plt.title('Box Blur', fontsize=18)

plt.show()

#Credit and reference: https://github.com/SamratSahoo/UT-Arlington-Research/blob/master/Week%206%20-%20Convolutions%20%26%20Wavelet%20Transforms/Convolutions.py
```

### <font color='#34A853'>Strides</font>

><font size="3">**When computing the <span style="color:#4285F4">feature map</span>, the convolution window starts at the upper-left corner of the input tensor, and then slide it over all locations both down and to the right.**</font>

><font size="3">**Sometimes, either for computational efficiency or downsampling, the convolution window may slide more than one element at a time, skipping the intermediate locations.**</font>

><font size="3">**To determine how many element it slides for each step, <span style="color:#4285F4">stride</span> is needed. <span style="color:#4285F4">Stride</span> is the <span style="color:#4285F4">number of pixels shifts</span> over the input matrix.**</font>
>><font size="3">**When the stride is 1 then we move the filters to 1 pixel at a time.**</font><br>
>><font size="3">**When the stride is 2 then we move the filters to 2 pixels at a time and so on.**</font>

### <font color='#34A853'>Padding</font>

><font size="3">**When applying convolutional layers, the pixels on the perimeter of an image are lost. As a result, features close to the perimeter cannot be captured properly. Since we typically use small kernels, for any given convolution, we might only lose a few pixels, but this can add up as we apply many successive convolutional layers.**</font>

><font size="3">**One straightforward solution to this problem is to <span style="color:#4285F4">add extra pixels around the perimeter</span> of an image so as to increase the effective size of the image, thus better capturing those features close to the perimeter.**</font>

<font size="3">**In TensorFlow2, the argument of padding can be one of <span style="color:#4285F4">"valid"</span> or <span style="color:#4285F4">"same"</span>.**</font> 
    
><font size="3">**<span style="color:#4285F4">"valid"</span> means <span style="color:#4285F4">no padding</span>.**</font> 
    
><font size="3">**<span style="color:#4285F4">"same"</span> results in <span style="color:#4285F4">padding with zeros</span> evenly to the left/right or up/down of the input <span style="color:#4285F4">such that output has the same height/width dimension as the input</span>.**</font>

### <font color='#34A853'>Activation</font>

><font size="3">**After computing the <span style="color:#4285F4">feature map</span>, the values will be transformed via an <span style="color:#4285F4">activation function</span>.**</font>

><font size="3">**<span style="color:#4285F4">Linear activation</span> is the simplest activation function, where no transform is applied at all. However, it cannot learn complex (nonlinear) mapping functions. <span style="color:#4285F4">Linear activation</span> functions are still used in the output layer for networks that predict a quantity (e.g. <span style="color:#4285F4">regression problems</span>).**</font>

><font size="3">**<span style="color:#4285F4">Nonlinear activation</span> functions are preferred as they allow the nodes to learn more complex structures in the data. Traditionally, two widely used nonlinear activation functions are:**</font>
>><font size="3">**<span style="color:#4285F4">Sigmoid</span>**</font><br>
>><font size="3">**<span style="color:#4285F4">Hyperbolic tangent (tanh)</span>**</font>


<font size="3">**The <span style="color:#4285F4">Sigmoid function</span> curve looks like a S-shape.**</font>

><font size="3">**The main reason why we use sigmoid function is because it exists between <span style="color:#4285F4">(0 to 1)</span>. Therefore, it is especially used for models where we have to predict the <span style="color:#4285F4">probability</span> as an output. Since probability of anything exists only between the range of 0 and 1, sigmoid is the right choice.**</font>

<font size="3">**A general problem with both the <span style="color:#4285F4">sigmoid</span> and <span style="color:#4285F4">tanh</span> functions is that they <span style="color:#4285F4">saturate</span>.**</font>
    
><font size="3">**This means that large values snap to <span style="color:#4285F4">1.0</span> and small values snap to <span style="color:#4285F4">-1 or 0</span> for <span style="color:#4285F4">tanh</span> and <span style="color:#4285F4">sigmoid</span> respectively.**</font>

><font size="3">**Further, the functions are only really <span style="color:#4285F4">sensitive to changes around their mid-point</span> of their input, such as 0.5 for <span style="color:#4285F4">sigmoid</span> and 0.0 for <span style="color:#4285F4">tanh</span>.**</font>

<font size="3">**<span style="color:#4285F4">Deep networks</span> using these nonlinear activation functions <span style="color:#4285F4">fail to receive useful gradient information</span>.**</font>
    
><font size="3">**Error is back propagated through the network and used to update the weights. The <span style="color:#4285F4">amount of error decreases dramatically</span> with each additional layer, given the derivative of the chosen activation function, namely <span style="color:#4285F4">vanishing gradient problem</span>. This prevents deep networks from learning effectively.**</font>

<font size="3">**<span style="color:#4285F4">To solve the vanishing gradient problem</span>, especially for deep networks, an activation function is needed that looks and acts like a linear function, but is, in fact, a nonlinear function allowing complex relationships in the data to be learned.**</font>

><font size="3">**A famous function is to use the <span style="color:#4285F4">rectified linear activation unit (ReLU)</span>.**</font>

<font size="3">**<span style="color:#4285F4">ReLU function</span> has <span style="color:#4285F4">gradient 1 when input > 0, and zero otherwise</span>.**</font>
    
><font size="3">**Thus, multiplying a bunch of ReLU derivatives together in the backprop equations has the nice property of being either 1 or 0.**</font>

><font size="3">**There is no <span style="color:#4285F4">vanishing</span> of the gradient. The gradient travels to the bottom layers either as is or it becomes exactly 0 on the way.**</font>

## 5.4 Pooling Layer

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN2.png?alt=media&token=9b2ffe9a-bbe2-40e9-986f-2f47ef012b91" width="750" align="center"/>

<font size="3">**<span style="color:#4285F4">Convolutional layers</span> prove very effective, and stacking convolutional layers in deep models allows layers close to the input to learn low-level features, like lines, and layers deeper in the model to learn high-order or more abstract features, like shapes or specific objects.**</font> 

<font size="3">**A limitation of the feature map output of convolutional layers is that they record the <span style="color:#4285F4">precise position of features</span> in the input. This means that small movements in the position of the feature in the input image will result in a different feature map. This can happen with re-cropping, rotation, shifting, and other minor changes to the input image.**</font> 


<font size="3">**A common way to address this problem is to add a <span style="color:#4285F4">pooling layer</span> after <span style="color:#4285F4">convolutional layer</span> so as to <span style="color:#4285F4">down sample</span> the features.**</font> 

<font size="3">**The pooling layers are used to**</font><br>
><font size="3">**<span style="color:#4285F4">reduce the dimensions of the feature maps</span>, hence the number of parameters to learn and the amount of computation performed in the network.**</font><br>

><font size="3">**<span style="color:#4285F4">summarize the features</span> present in a region of the feature map generated by a convolution layer. So, further operations are performed on summarized features so as to make the model more robust to variations in the position of the features in the input image, namely <span style="color:#4285F4">translation invariance</span>.**</font><br>

### <font color='Orange'> Pooling </font>

<font size="3">**The pooling operation is specified, rather than learned. Two common functions used in the pooling operation are:**</font> 

><font size="3">**<span style="color:#4285F4">Maximum Pooling (or Max Pooling)</span>** - calculate the maximum value for each patch of the feature map.</font><br>

><font size="3">**<span style="color:#4285F4">Average Pooling</span>** - calculate the average value for each patch on the feature map.</font><br>

### <font color='#34A853'>Maximum Pooling</font>

><font size="3">**<span style="color:#4285F4">Maximum Pooling (or Max Pooling)</span> is a pooling operation that selects the <span style="color:#4285F4">maximum element</span> from the region of the feature map covered by the <span style="color:#4285F4">filter</span>.**</font>
    
><font size="3">**Thus, the output after <span style="color:#4285F4">max pooling</span> layer would be a feature map containing <span style="color:#4285F4">the most prominent features</span> of the feature map.**</font>

<font size="3">**Let's assume a <span style="color:#4285F4">maximum pooling</span> with <span style="color:#4285F4">filters=(2,2)</span> and <span style="color:#4285F4">stride=(2,2)</span> is applied to a <span style="color:#4285F4">8x8 input</span>.**</font>
    
<font size="3">**For each <span style="color:#4285F4">max pool</span>, the operation computes the <span style="color:#4285F4">max value</span>. Doing so for each <span style="color:#4285F4">pool</span>, we will get a downsampled <span style="color:#4285F4">4Ã—4 output</span>.**</font>

<img src="https://firebasestorage.googleapis.com/v0/b/deep-learning-crash-course.appspot.com/o/5CNN16.gif?alt=media&token=c9f5600e-37ab-440f-9a5b-0b8864dd433e" width="430" style="float:center"/>

<font size="5"><span style="background-color:#EA4335; color:white">&nbsp;!&nbsp;</span></font> <font size="3">**The <span style="color:#4285F4">stride</span> is often equal to the <span style="color:#4285F4">pool size</span>, so that its effect equals a reduction in height and width (By a factor of 2 in width and 2 in height in our example).**</font>

### <font color='#34A853'>Average Pooling</font>

><font size="3">**<span style="color:#4285F4">Average Pooling</span> is a pooling operation that retains the <span style="color:#4285F4">average value</span> from the region of the feature map covered by the <span style="color:#4285F4">filter</span>.**</font>

><font size="3">**Thus, the output after <span style="color:#4285F4">average pooling</span> layer would be a feature map containing <span style="color:#4285F4">average values of features</span> of the feature map.**</font>

<font size="3">**Let's assume an <span style="color:#4285F4">average pooling</span> with <span style="color:#4285F4">filters=(2,2)</span> and <span style="color:#4285F4">stride=(2,2)</span> is applied to a <span style="color:#4285F4">8x8 input</span>.**</font>
    
<font size="3">**For each <span style="color:#4285F4">average pool</span>, the operation computes the <span style="color:#4285F4">average value</span>. Doing so for each <span style="color:#4285F4">pool</span>, we will get a downsampled <span style="color:#4285F4">4Ã—4 output</span>.**</font>

### <font color='#176BEF'> Examples </font>

<font size="3">**Keras provides <span style="color:#4285F4">1D</span> layer, <span style="color:#4285F4">2D</span> layer and <span style="color:#4285F4">3D</span> layer for <span style="color:#4285F4">MaxPooling</span> and <span style="color:#4285F4">AveragePooling</span>.**</font><br>

><font size="3">**<span style="color:#4285F4">1D pooling</span> layer is used with <span style="color:#4285F4">Conv1D</span> which is applied in previous layer.**</font><br>
><font size="3">**<span style="color:#4285F4">2D pooling</span> layer is used with <span style="color:#4285F4">Conv2D</span> which is applied in previous layer.**</font><br>
><font size="3">**<span style="color:#4285F4">3D pooling</span> layer is used with <span style="color:#4285F4">Conv3D</span> which is applied in previous layer.**</font><br>

<font size="3">**Commonly used arguments include:**</font><br>
    
><font size="3">**pool_size** - An integer or tuple of 2 integers, specifying the height and width of the pooling window. If only one integer is specified, the same window length will be used for both dimensions.</font><br>

><font size="3">**strides** - An integer or tuple of 2 integers or None, specifying how far the pooling window moves for each pooling step. If None, it will default to pool_size.</font><br>

><font size="3">**padding** - One of <span style="color:#4285F4">"valid"</span> or <span style="color:#4285F4">"same"</span></font><br>
>><font size="3"><span style="color:#4285F4">"valid"</span> means no padding.</font><br>
>><font size="3"><span style="color:#4285F4">"same"</span> results in padding with zeros evenly to the left/right or up/down of the input such that output has the same height/width dimension as the input.</font><br>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

<font size="3">**Let's assume we have a <span style="color:#4285F4">32x32 RGB</span> image.**</font><br>
    
<font size="3">**We want to create <span style="color:#4285F4">64 filters</span> with <span style="color:#4285F4">3x3 kernel size</span> which <span style="color:#4285F4">shift 1 pixel</span> at a time.**</font><br>

<font size="3">**The output has the same height/width dimension (i.e. <span style="color:#4285F4">32x32</span>) as the input and is activated by <span style="color:#4285F4">ReLU</span>.**</font><br>

```python
model = Sequential()
model.add(Conv2D(filters = 64, 
                 kernel_size = (3, 3), 
                 strides = (1, 1), 
                 padding = "same", 
                 activation = 'relu', 
                 input_shape = (32, 32, 3)))
model.summary()
```

<font size="3">**Let's assume a maximum pooling with filters=(2,2) and stride=(2,2) is applied to a 32x32 output from previous <span style="color:#4285F4">Conv2D</span> to downsample.**</font><br>

```python
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid'))
model.summary()
```

<font size="3">**Doing so for each <span style="color:#4285F4">pool</span>, we will get a downsampled <span style="color:#4285F4">16x16 output</span>.**</font>

## 5.5 Flatten Layer

<font size="3">**Once the model has learnt the features, it is ready to flatten the final feature map and feed it to a neural network (e.g. few dense layers) for classification purposes.**</font>

<font size="3">**However, <span style="color:#4285F4">dense layers</span> take <span style="color:#4285F4">vectors</span> as input (which are 1D), while the current output is a <span style="color:#4285F4">3D tensor</span>.**</font>
    
<font size="3">**Therefore, a <span style="color:#4285F4">flatten layers</span> is used to convert the data into a 1-dimensional array for inputting it to the next layer.**</font>

<font size="3">**E.g., a 3x3 image matrix into a 9x1 vector**</font>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

<font size="3">**The output has the same height/width dimension (i.e. <span style="color:#4285F4">32x32</span>) as the input and is activated by <span style="color:#4285F4">ReLU</span>.**</font><br>

<font size="3">**Then, a maximum pooling with filters=(2,2) and stride=(2,2) is applied to a 32x32 output from previous <span style="color:#4285F4">Conv2D</span> to downsample.**</font><br>

```python
model = Sequential()
model.add(Conv2D(filters = 64, 
                 kernel_size = (3, 3), 
                 strides = (1, 1), 
                 padding = "same", 
                 activation = 'relu', 
                 input_shape = (32, 32, 3)))
model.add(MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid'))
model.summary()
```

## 5.6 Dense Layer

<font size="3">**<span style="color:#4285F4">Dense layer</span> is a <span style="color:#4285F4">fully-connected layer</span>. To complete the model, one or more <span style="color:#4285F4">dense layers</span> can be added.**</font>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

<font size="3">**Let's assume we have a <span style="color:#4285F4">32x32 RGB</span> image.**</font><br>
    
<font size="3">**We create <span style="color:#4285F4">64 filters</span> with <span style="color:#4285F4">3x3 kernel size</span> which <span style="color:#4285F4">shift 1 pixel</span> at a time.**</font><br>

<font size="3">**The output has the same height/width dimension (i.e. <span style="color:#4285F4">32x32</span>) as the input and is activated by <span style="color:#4285F4">ReLU</span>.**</font><br>

<font size="3">**A maximum pooling with filters=(2,2) and stride=(2,2) is then applied to a 32x32 output from previous <span style="color:#4285F4">Conv2D</span> to downsample.**</font><br>
